```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets("mnist_data/",one_hot=True)
```

```python
training_digits, training_labels = mnist.train.next_batch(5000)
test_digits, test_labels = mnist.test.next_batch(200)
```

```python
training_digits_pl = tf.placeholder("float",shape=[None,784])
```

```python
test_digits_pl = tf.placeholder("float",shape=[784])
```

```python
l1_distance = tf.abs(tf.add(training_digits_pl,tf.negative(test_digits_pl)))
```

```python
distance = tf.reduce_sum(l1_distance,axis=1)
```

```python
pred = tf.argmin(distance,0)
```

```python
init = tf.global_variables_initializer()
```

```python
accuracy = 0.
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    for i in range(len(test_digits)):
        nn_index = sess.run(pred, feed_dict={training_digits_pl: training_digits, test_digits_pl:test_digits[i,:]})
        print("Test", i, "Prediction: ", np.argmax(training_labels[nn_index]), \
              "True Label: ", np.argmax(test_labels[i]))
        if np.argmax(training_labels[nn_index]) == np.argmax(test_labels[i]):
            accuracy += 1./len(test_digits)
    print("Done!")
    print("Accuracy: ",accuracy)
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets("mnist_data/",one_hot=True)
```

```python
train_data, train_labels = mnist.train.next_batch(500)
test_data, test_labels = mnist.test.next_batch(200)
```

```python
training_data_pl = tf.placeholder(dtype=tf.float32,name='training_data_pl',shape=[None,784])
```

```python
test_data_pl = tf.placeholder(dtype=tf.float32,name='test_data_pl',shape=[784])
```

```python
l1_distance = tf.abs(tf.add(training_data_pl,tf.negative(test_data_pl)))
```

```python
distance = tf.reduce_sum(l1_distance,axis=1)
```

```python
pred = tf.arg_min(distance,0)
```

```python
init = tf.global_variables_initializer()
```

```python
accuracy = 0
with tf.Session() as sess:
    sess.run(init)
    for i in range(len(test_data)):
        index = sess.run(pred,feed_dict={training_data_pl:train_data,test_data_pl:test_data[i,:]})
        print("Test", i, "Prediction: ", np.argmax(train_labels[index]), \
        "True Label: ", np.argmax(test_labels[i]))
        if np.argmax(train_labels[index]) == np.argmax(test_labels[i]):
            accuracy += 1./len(test_data)
    print("Done!")
    print("Accuracy: ",accuracy)
```
